# Sequential Uncertainty Reuction and Value of Information in Gaussian Process Models

Consider the situation where a farmer must decide wether to harvest a forest with uncertain profits, or just to leave it for conservation and receive 0 profits.

The Forest: $25 \times 25$ grid, consisting of 625 units of cells $(1,1),(1,2),\dots,(25,25)$.

<img src="https://i.imgur.com/zqlPWhR.png" alt="grid" style="display: block; margin-left: auto;margin-right: auto;width: 80%;">

* $a=0$: leave all cells for conservation
* $a=1$: harvest all cells

Cells:    $\mathbf{s} = (s_1,s_2)$

The profit in each cell: $ \mathbf{x} = (x_1,\dots,x_{625})$

Expected profit: $\mathbf{E}[x(\mathbf{s})] = \mathbf{0}$

Variance of profit: $\mathrm{Var}[x(\mathbf{s})] = 1^2 $

Covariance of profit between cells: $\mathrm{Cov}[x(\mathbf{s}),x(\mathbf{s'})] = \exp(-0.15h), \enspace \textrm{where } h = \sqrt{||\mathbf{s}-\mathbf{s}'||^2}$

Gives the prior of profit: $\mathbf{x} \sim \mathcal{N}[\mathbf{0},\mathbf{\Sigma}]$

<img src="https://i.imgur.com/gHwoEN3.png" alt="sample grid" style="display: block; margin-left: auto;margin-right: auto;width: 80%;">

Each cell sampled with: $y(\mathbf{s}) = x(\mathbf{s}) + \mathcal{N}[0,5^2]$

The $j$th line sampled with(the likelihood): $\mathbf{y}_j|\mathbf{x} \sim \mathbf{F}_j\mathbf{x} + \mathcal{N}[\mathbf{0},\mathbf{T}] = \mathcal{N}[\mathbf{F}_j\mathbf{x},\mathbf{T}]$

$\mathbf{F}_j$ is of size $25 \times 625$ picks entries from $\mathbf{x}$

new mean : $\boldsymbol{\mu}_j = \boldsymbol{\mu} + \mathbf{\Sigma}\mathbf{F}_j^T(\mathbf{F}_j\mathbf{\Sigma}\mathbf{F}_j^T + \mathbf{T})^{-1}(\mathbf{y}_j - \mathbf{F}_j\boldsymbol{\mu})$

new covariance: $\mathbf{\Sigma}_j =  \mathbf{\Sigma} - \mathbf{\Sigma}\mathbf{F}_j^T(\mathbf{F}_j\mathbf{\Sigma}\mathbf{F}_j^T + \mathbf{T})^{-1}\mathbf{F}_j\mathbf{\Sigma}$

```python
# imports
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [8.0, 8.0]
import matplotlib.pyplot as plt
from scipy.stats import norm # importing norm cdf and pdf
import pandas as pd # to plot dictionaries
import matplotlib.ticker as ticker
```

```python
N = 25
m = 625
def create_grid(m,N):
    Sigma = np.identity(m)
    s = np.zeros((2,m))
    T = np.identity(N)*5**2
    mu = np.zeros(m)
    for i in range(N):
        for j in range(N):
            s[:,j + i*N] = [i+1,j+1]
    for i in range(m-1):
        for j in range(i+1,m):
            Sigma[i,j] = Sigma[j,i] = np.exp(-0.15*np.sqrt((s[0,i] - s[0,j])**2 + (s[1,i] - s[1,j])**2))
    return s, mu, Sigma, T
```

```python
s, mu, Sigma, T = create_grid(m,N)
```

```python
plt.matshow(Sigma);
```

```python
plt.matshow(T);
```

```python
def get_F_j(m,N,j):
    F_j = np.zeros((N,m))
    for i in range(N):
        for k in range(N):
            F_j[i,k*N + j-1] = 1
    return F_j

def get_Sigma_j(m,N,j,Sigma,T):
    F_j = get_F_j(m,N,j)
    tmp = (F_j@Sigma@F_j.transpose() + T)
    tmp = np.linalg.inv(tmp)
    Sigma_j = Sigma - Sigma@F_j.transpose()@tmp@F_j@Sigma
    return Sigma_j, F_j
```

```python
F_j = get_F_j(m,N,1)
plt.matshow(F_j);
```

## Select the line $j$ that maximises the reduction in the mean variance of the grid

$$j^* = \underset{j}{\mathrm{argmin}}\frac{1}{625}\sum_{k=1}^{625} \Sigma_j(k,k)$$

```python
def max_reduc_mean_var(m,N,Sigma,T):
    tmp_min = np.inf
    tmp_min_j = np.nan
    for j in range(1,26):
        Sigma_j, F_j = get_Sigma_j(m,N,j,Sigma,T)
        tmp_mean_var = np.diag(Sigma_j).mean()
        if tmp_mean_var < tmp_min:
            tmp_min = tmp_mean_var
            tmp_min_j = j
    return tmp_min_j
```

```python
min_j= max_reduc_mean_var(m,N,Sigma,T)
print("Line {} maximizes the reduction in mean variance".format(min_j))
```

## Conduct sequential selecton of lines

$$\mathbf{\Sigma}_j = \mathbf{\Sigma} - \mathbf{\Sigma}\mathbf{F}_j^T(\mathbf{F}_j\mathbf{\Sigma}\mathbf{F}_j^T + \mathbf{T})^{-1}\mathbf{F}_j\mathbf{\Sigma}$$

### Find the 10 first lines:

```python
def ten_lines(m,N):
    s, mu, Sigma, T = create_grid(m,N)
    for i in range(10):
        min_j = max_reduc_mean_var(m,N,Sigma,T)
        Sigma_j, F_j = get_Sigma_j(m,N,min_j,Sigma,T)
        Sigma = Sigma_j
        print("Line {} maximizes the reduction in mean variance".format(min_j))
    return mu, Sigma
```

```python
mu, Sigma = ten_lines(m,N)
```

## Decision making: harvest ($a=1$), or not to harvest ($a=0$)

Prior value: $$PV = \max_{a\in A} \{\mathbf{E}[v(\mathbf{x},a)]\}$$

$$ = \max\left\{0,\sum_{k=1}^{625} \mu_k\right\}$$

Posterior value: $$ PoV (\mathbf{y}) = \int \max_{a\in A} \left\{\mathbf{E}[v(\mathbf{x},a)|\mathbf{y}] \right\}p(\mathbf{y})\mathrm{d}\mathbf{y}$$

$$ = \sum_y \max_{a\in A} \left\{\mathbf{E}[v(\mathbf{x},a)|y)\right\}p(y)$$

$$ = \mu_w\Phi(\mu_w/r_{w,j}) +  r_{w,j}\phi(\mu_w/r_{w,j})$$

Where: 
$$\mu_w = \sum_i \mu_i$$
$$\mathbf{R}_j =\mathbf{\Sigma}\mathbf{F}_j^T(\mathbf{F}_j\mathbf{\Sigma}\mathbf{F}_j^T + \mathbf{T})^{-1}\mathbf{F_j}\mathbf{\Sigma}$$
$$r_{w,j} = \sqrt{\sum_k\sum_l R_j(k,l)}$$
$$ \Phi(\cdot)\enspace -  \textrm{ the standard normal CDF}$$
$$ \phi(\cdot)\enspace -  \textrm{ the standard normal PDF}$$

Value of Information: $$VOI(\mathbf{y}) = PoV(\mathbf{y}) - PV $$

### Myopic Scheme

\begin{equation*}
\boxed{\begin{array}{ll}
    \textrm{Initiate:} & \mathbf{x}, \boldsymbol{\mu}, \mathbf{\Sigma}\\ 
    \textrm{Find:} & j^* = \underset{j}{\mathrm{argmin}} \frac{1}{m}\sum_{k=1}^m \Sigma_j(k,k)\\
    \textrm{Calc:} & \mathbf{R}_{j^*}, \mu_w, r_{w,j^*}, \mathrm{VOI}_{j^*}\\
    \textrm{While }\mathrm{VOI}_{j^*} > P:  &  \\ 
    \hspace{25pt}\textrm{Collect data:} & \mathbf{y}_{j^*}\\
    \hspace{25pt}\textrm{Update model:} & \boldsymbol{\mu} = \boldsymbol{\mu}+\mathbf{\Sigma}\mathbf{F}_{j^*}^T(\mathbf{F}_{j^*}\mathbf{\Sigma}\mathbf{F}_{j^*}^T + \mathbf{T})^{-1}(\mathbf{y}_{j^*} - \mathbf{F}_{j^*}\boldsymbol{\mu}) \\ 
    & \mathbf{\Sigma} \mathrel{{-}{=}} \mathbf{R}_{j^*}\\ 
    \hspace{25pt}\textrm{Find:} & j^* = \underset{j}{\mathrm{min}} \frac{1}{m}\sum_{k=1}^m \Sigma_j(k,k)\\
    \hspace{25pt}\textrm{Calc:} & \mathbf{R}_{j^*}, \mu_w, r_{w,j^*}, \mathrm{VOI}_{j^*}\\
\end{array}}
\end{equation*}

```python
def sample_y_j(mu,Sigma,F_j,T):
    return np.random.multivariate_normal(F_j @ mu, 
                                         F_j @ Sigma @ F_j.transpose() + T)
```

```python
# function calculating the value of information
def value_of_information(mu_w,r_w,PV):
    return mu_w*norm.cdf(mu_w/r_w) + r_w*norm.pdf(mu_w/r_w) - PV
```

```python
# code of the myopic scheme
def sequential_voi(m,N):
    s, mu, Sigma, T = create_grid(m,N)
    min_j = max_reduc_mean_var(m,N,Sigma,T)
    Sigma_j, F_j = get_Sigma_j(m,N,min_j,Sigma,T)
    mu_w = mu.sum()
    r_w = np.sqrt((Sigma-Sigma_j).sum())
    PV = 0
    voi = value_of_information(mu_w,r_w,PV)
    # creating ouput lists
    lines = []
    voi_list = []
    n_lines = 0
    while voi > 0.5:
        lines.append(min_j)
        voi_list.append(voi)
        n_lines += 1
        y_j = sample_y_j(mu,Sigma,F_j,T)
        mu = mu + Sigma @ F_j.transpose()@np.linalg.inv(T + F_j@Sigma@F_j.transpose())@(y_j-F_j@mu)
        Sigma = Sigma_j
        PV = np.max([mu.sum(),0])
        min_j = max_reduc_mean_var(m,N,Sigma,T)
        Sigma_j, F_j = get_Sigma_j(m,N,min_j,Sigma,T)
        mu_w = mu.sum()
        r_w = np.sqrt((Sigma-Sigma_j).sum())
        voi = value_of_information(mu_w,r_w,PV)
    return lines, int(n_lines),voi_list
```

```python
lines, n_lines,voi = sequential_voi(m,N)
print("Example run selected {} lines, and the lines where {}".format(n_lines,lines))
```

## 100 replicated trials

```python
rep_trials = 100
lines_df = pd.DataFrame({"lines":[],
                        "trial":[],
                        "voi":[]})
for i in range(rep_trials):
    lines, n_lines, voi = sequential_voi(m,N)
    lines_df = lines_df.append(
        pd.DataFrame({"lines":lines,
                      "trial":[i+1]*int(n_lines),
                      "voi":voi}),
        ignore_index = True)
    print("Trial {}: {} lines".format(i+1,n_lines),end="\r")
```

```python
g = sns.jointplot(x="trial", y="lines", data=lines_df,marginal_kws=dict(bins=100),height=10, ratio=3, color="b",annot_kws=dict(stat="r"), edgecolor="w", linewidth=1)
g.ax_joint.yaxis.set_major_locator(ticker.MultipleLocator(1))
```

The figure above shows the different replicated trials along the x-axis with their respective selection of lines along the y-axis. The histogram on top shows the number of selected lines for each trial, and the histogram to the right shows the histogram of number of times the a line is chosen in all the trials.

```python
ax = sns.distplot(lines_df["trial"], bins = 100,kde=False)
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
```

The figure above presents the histogram also included in the previous figure. It displays the number of selected lines in each trial.

We can see that line 13 is chosen every time, followed by the most occuring is line 22 and 4. Lines 18 and 9 has also been chosen some trials. Also occuring is line 25 and 1. Lines 7 and 15 have occured ones or twice. In the last plot we can see that we have at most chosen 9 lines. 

The Myopic scheme as previously mentioned is a greedy scheme, and it doesn't look beyond the next step to evaluate if more information is good. If we were to look at 10 runs for every trial and then look at the VOI in the later steps, this might be higher than $0.5$.
